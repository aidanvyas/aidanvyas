### Python Lists and Dictionaries

Lists and dictionaries are essential in Python. You need to know how to use them.



Let's explore **lists**

```
mylist = ["cat", "dog", "zebra", "elephant"] # You define a list like this.
print(mylist)  # You can print the list directly.

print(mylist[0])  # You access an item in the list like this.
mylist[0] = "kitty"  # You modify an item in the list like this.
print(mylist)

mylist.append("bear")  # You add an element ot the list like this.
print(mylist)

print(len(mylist)) # You print the length of a list like this.

# This is how you iterate through a list.
for (i, element) in enumerate(mylist):
  print("mylist[{0}] = {1}".format(i, element))
```

Now let's see what we can do with **dictionaries**.



```
# You define a dictionary like this.
mydictionary = {"Andrew": 26, "Damon": 22, "Yunjie": 21,
                "Jerome": 20, "Courtney": 21}

print(mydictionary) # You can just print the dictionary like this.
print(mydictionary.keys()) # You can also print the keys.
print(mydictionary.values())  # Or you can print the values.

# You can access the values of the dictionary using the keys.
print(mydictionary["Courtney"])

# You can change the values in the dictionary like this.
mydictionary["Courtney"] = 19
print(mydictionary)

# You can traverse a dictionary like this.
for (key, value) in mydictionary.items():
  print("mydictionary[{0}] = {1}".format(key, value))

# You can also check if a key is in the dictionary like this.
print("Courtney" in mydictionary)
print("Vicente" in mydictionary)
```

**More Advanced:** List/Dictionary comprehensions. This is something you ultimately want to get to. Being good at writing one-liners with list comprehensions. I give some examples here of things you can do with Python. If you master how to do these you will be able to write a lot less code.

```
# Here I'm showing you how to create a list with even numbers from 0-19.
even_list = [x for x in range(0, 20) if x % 2 == 0]
print(even_list)

# Here I'm creating a dictionary that maps every even number in the previous
# list to their square version.
square_evens = {x: x ** 2 for x in even_list}

print(square_evens)
```

### Numpy Arrays

Numpy is a powerful library for linear algebra. It allows to define scalar, vectors, matrices, and n-dimensional arrays (also referred as tensors which are just matrices with more than two dimensions [For the purposes of this class]).

```
import numpy as np

# Let's define a matrix below.
my_matrix = np.array([[3, 2, 2],
                      [2, 5, 2],
                      [4, 4, 7]])

print(my_matrix) # You can show it like this.
print(my_matrix.shape) # You can print the shape of the matrix.
# You can also get the number of rows and columns like this.
print("This matrix is {0}x{1}".format(my_matrix.shape[0], my_matrix.shape[1]))


# We could traverse the matrix as follows.
for (i, row) in enumerate(my_matrix):
  for (j, value) in enumerate(row):
    print("m[{0},{1}] = {2}".format(i, j, value))

# We could also traverse the matrix as follows
for i in range(0, my_matrix.shape[0]):
  for j in range(0, my_matrix.shape[1]):
    print("m[{0},{1}] = {2}".format(i, j, my_matrix[i, j]))

# HOWEVER: MOST OF THE TIMES YOU DON'T NEED TO TRAVERSE A MATRIX.
# TRAVERSING A MATRIX WILL MAKE YOUR CODE SLOW. YOU SHOULD AVOID TRAVERSING.
```

### Indexing and Slicing Arrays

There are many different ways to either retrieve or manipulate only part of the array. This is an extensive topic and it is done through indexing and slicing. I recommend you look at the numpy documentation for this part but I include some examples here https://docs.scipy.org/doc/numpy/reference/arrays.indexing.html

```
# Let's define a matrix below.
my_matrix = np.array([[3, 2, 2],
                      [2, 5, 2],
                      [4, 4, 7]])

# Basic indexing.
# Let's print an element in the matrix.
print(my_matrix[1,1])

# Basic slicing.
# Let's print the values in between positions 1:3 and 1:3 (3 not inclusive)
print(my_matrix[1:3, 1:3]) # This should print [[5,2], [4,7]].

# Let's print only the first row.
print(my_matrix[0, :])  # This should print [[3,2,2]]
# Equivalent to print(my_matrix[0, 0:3]) where 3 is number of columns.

# Let's print only the first column.
print(my_matrix[:, 0]) # This should print [[3,2,4]]
# Equivalent to print(my_matrix[0:3, 0]) where 3 is number of rows.

# Let's print only the top-left 2x2 values in this 3x3 matrix.
print(my_matrix[:2, :2]) # This should print [[3,2],[2,5]]
# Equivalent to print(my_matrix[0:2,0:2])
```

Here I present more advanced indexing. Being good at indexing and slicing arrays will be tremendously helpful to avoid matrix traversals.

```
# Logical indexing: Here we access or modify elements of the array
# where a condition is met.

# First, realize we can make logical operations on a numpy array.
# This should print a matrix with True or False in each position
# depending on the value of the logical statement.
print(my_matrix > 3)

# We can use this matrix to index on the original array!
# This prints a list with values that are bigger than 3 in the original array.
out_matrix = my_matrix[my_matrix > 3]
print(out_matrix)

# IMPORTANT: Make a copy of the array using np.array
# Without this, the second array variable will not be a copy,
# i.e. modifying it would still change the original matrix.
my_matrix_copy = np.array(my_matrix)
my_matrix_copy[my_matrix_copy > 3] = 0
print(my_matrix_copy) # Values bigger than 3 are set to zero in the copy.
print(my_matrix) # Original matrix still has the same values.
```

More advanced indexing can be achieved using some of the pre-defined indexing utility functions in numpy such as **numpy.take** (https://docs.scipy.org/doc/numpy/reference/generated/numpy.take.html), **numpy.indices** (https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.indices.html#numpy.indices), **numpy.where** (https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.where.html#numpy.where), **numpy.nonzero** (https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.nonzero.html#numpy.nonzero). These are among important ones. But please check a list of all of them here https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.indexing.html.

Most of the times for-loops can be avoided by using one of those indexing functions from numpy.

### Basic Numpy Array Operations.

Most of the times you don't need to traverse a matrix. Traversing a matrix is slow. Numpy provides many many functions to avoid traversing a matrix. I give some examples here. You can find examples for almost any operation you can think of without a need to ever traverse a matrix. Most operations you need to do in an image they can be done without using matrix traversal with for-loops.

```
# Matrix-scalar multiplication.
scaled_matrix = my_matrix * 1.33
print(scaled_matrix)

# Matrix element-wise square.
squared_matrix = my_matrix ** 2
print(squared_matrix)

# Matrix element-wise square-root.
sqrt_matrix = np.sqrt(my_matrix)
print(sqrt_matrix)

# Matrix transpose
transp_matrix = my_matrix.T
print(transp_matrix)

# Matrix re-shaping into vector.
vectorized_matrix = my_matrix.reshape((9, 1))
print(vectorized_matrix)
```

### Array-Array Numpy Operations

Sometimes you will need to do matrix-matrix multiplication and other operations between vectors or matrices. Here are some examples.

```
# Let's define two matrices below.
A = np.array([[3, 2, 2],
              [2, 5, 2]])

B = np.array([[1, 2, 1],
              [3, 3, 1]])

# We can add the two matrices.
print(A + B)

# We can element-wise multiply the values in the two matrices.
print(A * B)

# However matrix-matrix multiplication between A and B is not possible
# because of their shapes. They are both 2x3 matrices.
# So we need to transpose one to make matrix-matrix multiplication possible.
# Here I show two ways to multiply them:

print(np.matmul(A, B.T)) # A is 2x3 and B-transpose is 3x2 -> output 2x2
print(np.matmul(A.T, B)) # A-transpose is 3x2 and B is 2x3 -> output 3x3

# We could also use np.dot instead of np.matmul. See the documentation
# for np.dot https://numpy.org/devdocs/reference/generated/numpy.dot.html
```

**Advanced:** If you want to take your numpy operations between n-dimensional arrays to the next level I recommend you master Broadcasting. This allows you to operate between arrays that are seemingly incompatible for the operation. Numpy will automatically resize or reshape the arrays for you but you have to understand the behavior of numpy -- so you have to be careful. The Stanford's Python-Numpy tutorial provides a more comprehensive detail on Broadcasting http://cs231n.github.io/python-numpy-tutorial/. Broadcasting is the ultimate way to avoid matrix traversals to the maximum extent possible.

### Instagram Primer

In this section I will show you some cheap implementation in few lines of code for Instagram operations.

```
import numpy as np
import imageio
import matplotlib.pyplot as plt

img_url = 'http://www.cs.virginia.edu/~vicente/images/google_android.jpg'
img = imageio.imread(img_url) / 255.0
print(img.shape)

# Plot the image.
def display_image(img):
  plt.figure(); plt.axis('off'); plt.imshow(img); plt.grid(False); plt.show()

display_image(img)
```

#### 1. Cropping

```
# Cropping myself in the above image.
# You can do this in one line by using indices along the first and second
# dimensions (HEIGHT and WIDTH) and using the implicit indexer ":" for
# the third dimensio since we need all three channels.
display_image(img[172:240, 305:355, :])
```

#### 2. Warmth Adjustment.

```
# This is an example of broadcasting, where Numpy figures out that the second
# matrix is a vector of size 3, and the first matrix is the image of size
# 416x600x3, since the second vector matches along the third dimension, Numpy
# automatically replicates this vector along the other two dimensions so that
# the two matrices can be element-wise multiplied.
display_image(img)
display_image(img * [1, 0.85, 0.75])
```

#### 3. Saturation

```
from skimage.color import rgb2hsv, hsv2rgb

# Convert to HSV.
hsv = rgb2hsv(img)
# Enhance the S channel by a factor of 3.
# We need to clip so that pixels are still between 0 and 1.
hsv[:, :, 1] = np.clip(hsv[:, :, 1] * 3, 0, 1.0)
# Convert back to RGB.
saturated_img = hsv2rgb(hsv)

display_image(img)
display_image(saturated_img)
```

#### 4. Fade

```
# Fading means you interpolate every pixel in the image
# with a gray-colored pixel. This example uses broadcasting and
# scalar-matrix multiplication.
display_image(img)
alpha = 0.45
display_image(img * (1.0 - alpha) + [0.5 * alpha, 0.5 * alpha, 0.5 * alpha])
```

#### 5. Structure

```
from scipy.signal import convolve2d

sharpening_kernel = [[-1, -1, -1],
                     [-1,  9, -1],
                     [-1, -1, -1]]

# Sharpen each channel R, G, and B.
sharpened_r = convolve2d(img[:, :, 0], sharpening_kernel)
sharpened_g = convolve2d(img[:, :, 1], sharpening_kernel)
sharpened_b = convolve2d(img[:, :, 2], sharpening_kernel)

# Put the channels together again along the third dimension.
# Check here the usage of concatenate and np.newaxis.
# sharpened_r, sharpened_g, sharpened_b have only HxW dimensions.
# but we want to concatenate into the Channel dimension which they don't have.
# So we add a dummy dimension at the end so these arrays become HxWx1
# Then we can concatenate along the third dimension to get an array HxWx3.
sharpened = np.concatenate((sharpened_r[:, :, np.newaxis],
                            sharpened_g[:, :, np.newaxis],
                            sharpened_b[:, :, np.newaxis]), axis = 2)

display_image(img)
display_image(np.clip(sharpened, 0.0, 1.0))
```

#### 6. Contrast

```
# Contrast enhancement is also easy to implement in a single line of code.
# I took this equation from the GIMP book which implements all kinds of image
# processing routines http://pippin.gimp.org/image_processing/chap_point.html

contrast = 2
contrast_enhanced_img = np.clip((img - 0.5) * contrast + 0.5, 0.0, 1.0)

display_image(img)
display_image(contrast_enhanced_img)
```

#### 7. Jitter

```
# This is not from Instagram but from GIMP, I took the idea from the GIMP book
# http://pippin.gimp.org/image_processing/chap_area.html#id2560059
# Jitter replaces every pixel with a random pixel in its neighborhood.
# I don't know how to implement this without traversing the matrix :(
import random

(H, W, C) = img.shape # Yes, you can do this.

# Creates an image of size img.shape but initialized with zeros.
jitter_img = np.zeros_like(img)

# Traversing a matrix! So slow!
for i in range(0, H):
  for j in range(0, W):

    # Random offset but making sure indices don't go out of bounds.
    ii = np.clip(i + random.randint(-5, 5), 0, H - 1)
    jj = np.clip(j + random.randint(-5, 5), 0, W - 1)

    # Assign pixel value to current location.
    jitter_img[i, j, :] = img[ii, jj, :]

display_image(jitter_img)

# Well maybe I just thought of a way of avoiding traversals but we can
# discuss in class if you're curious.
```

#### 8. Green-in
This is also not from Instagram or GIMP, I just wanted to show how one could use logical indexing to perform a thresholding operation that only lets in the green color in this image.

```

# First I obtain a map of all the places where there is green color.
# I'm defining green color as Red < 0.3, Green > 0.5, Blue < 0.3
green_locations = (img[:, :, 0] < 0.3) * (img[:, :, 1] > 0.5) * (img[:, :, 2] < 0.3)

# Green locations contains 1 if there is green at the location, or 0 if not.
# So all I need to do is elementwise multiplication.
# However, img is 416x600x3 and green_locations is 416x600.
# I need to make them the same size. Here is one approach:
# I can use numpy.repeat https://docs.scipy.org/doc/numpy/reference/generated/numpy.repeat.html
# This function can repeat the green_locations along the third dimension,
# so that it can become 416x600x3.
green_locations = np.repeat(green_locations[:, :, np.newaxis], 3, axis = 2)
green_only_img = img * green_locations

# IMPORTANT: BROADCASTING SOLUTION IN ONE LINE OF CODE!
# I just add another dimension to green_locations
#  so that broadcasting does its job!
# Uncomment this line of code and it should work.
#green_only_img = img * green_locations[:, :, np.newaxis]

display_image(green_only_img)
```
